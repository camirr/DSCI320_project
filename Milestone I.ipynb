## Data Prepocressing


#### 1) Loading Dataset

```python
# Load packages
import pandas as pd
import altair as alt
import ast
```

```python
# Load dataset
url = 'https://raw.githubusercontent.com/shlrley/amazon_bestsellers/main/data/amazon_conlit_goodreads_nyt.csv'
amazon_book_original = pd.read_csv(url,parse_dates=['amazon_year'])
amazon_book_original.head()
```

```python
# Print out the column names, their data types, and count of non-NA values
amazon_book_original.info()
```

```python
# Find the number of unique values in each column
cardinality = amazon_book_original.nunique() 
print(cardinality)
```

```python
# Count the number of missing values for each column
amazon_book_original.isnull().sum()
```

#### 2) Data Wrangling

```python
# Drop the unused columns
amazon_book = amazon_book_original.drop(columns=['amazon_year', 'conlit_pubdate','conlit_author_nationality', 'conlit_total_ratings',
                                                 'goodreads_rating', 'goodreads_series', 'conlit_genre', 'goodreads_edition','goodreads_publisher',
                                                 'goodreads_publish_date', 'goodreads_first_publish_date', 'goodreads_awards', 'goodreads_num_ratings','goodreads_likedPercent',
                                                 'goodreads_price', 'nyt_published_date','nyt_list_name_encoded','nyt_price', 'conlit_author_gender']) 

#Remove missing values
amazon_book = amazon_book.dropna()


# Check if the dataset still contains any missing value
amazon_book.isnull().values.any()
```

```python
# Print out the column names, their data types, and count of non-NA values of cleaned dataset
amazon_book.info()
```

#### 3) Univariate Summaries

```python
# Univariate Numerical Summary - Quantitative
amazon_book.agg(
    {
    "amazon_rating": ['min', 'max', 'mean', 'median', 'std'],
    "amazon_num_reviews": ['min', 'max', 'median', 'std'],
    "amazon_price": ['min', 'max', 'median', 'std'],
    "nyt_weeks_on_list": ['min', 'max', 'median', 'std']
    }
)
```

```python
# Univariate Numerical Summary - Categorical
cat_attr = ['title', 'amazon_author', 'amazon_genre', 'goodreads_genres']

for n in cat_attr:
    print(n)
    print(pd.crosstab(index = amazon_book[n], columns = 'count'))
    print("\n")
```

```python
# Categorical Univariate Visual Summaries

alt.Chart(amazon_book).mark_bar().encode(
    y = alt.Y("title", sort = "-x"),
    x = 'count()'
)
```

```python
# Categorical Univariate Visual Summaries
alt.Chart(amazon_book).mark_bar().encode(
    y = alt.Y("amazon_author", sort = "-x"),
    x = 'count()'
)
```

```python
# Categorical Univariate Visual Summaries
alt.Chart(amazon_book).mark_bar().encode(
    y = alt.Y("amazon_genre", sort = "-x"),
    x = 'count()'
)
```

```python
# Categorical Univariate Visual Summaries
alt.Chart(amazon_book).mark_bar().encode(
    y = alt.Y("goodreads_genres", sort = "-x"),
    x = 'count()'
)
```

```python
# Convert columns with lists (but that are string type) to list type
amazon_book['goodreads_genres'] = amazon_book.goodreads_genres.apply(lambda s: list(ast.literal_eval(s)))
```

#### 4) Multivariate Summaries

```python
# Multivariate Numerical Summaries - amazon_rating & amazon_genre

pd.crosstab(index = amazon_book["amazon_rating"], columns = amazon_book["amazon_genre"])
```

```python
# Multivariate Numerical Summaries - amazon_price & title

pd.crosstab(index = amazon_book["amazon_price"], columns = amazon_book["title"])
```

```python
# Multivariate Visual Summaries - Bivariate Outlier Exploration
alt.Chart(amazon_book).mark_circle().encode(
    alt.X('amazon_price:Q'),
    alt.Y('count()'),
    alt.Color('amazon_genre'),
)
```

```python
# Multivariate Visual Summaries - Scatter Matrix
alt.Chart(amazon_book).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
).properties(
    width=100,
    height=100
).repeat(
    row=['amazon_rating', 'amazon_num_reviews', 'amazon_price', 'nyt_weeks_on_list'],
    column=['amazon_rating', 'amazon_num_reviews', 'amazon_price', 'nyt_weeks_on_list'],
)
```

```python
# Multivariate Visual Summaries - Bar Charts for nyt_weeks_on_list and amazon_author
alt.Chart(amazon_book).mark_bar().encode(
    alt.X('amazon_author:N', sort='-y'),
    alt.Y('sum(nyt_weeks_on_list):Q'),
    alt.Color('amazon_genre')
).interactive()
```

```python
# Multivariate Visual Summaries - Overlapping Density Plots for amazon_genre and amazon_num_reviews
alt.Chart(amazon_book).transform_density(
    'amazon_num_reviews',
    groupby=['amazon_genre'],
    as_=['amazon_num_reviews', 'density'],
).mark_area().encode(
    x="amazon_num_reviews:Q",
    y='density:Q',
).facet(
    'amazon_genre:N',
    columns=1
)
```
